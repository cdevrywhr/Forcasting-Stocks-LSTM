```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
antmk = pd.read_excel("ANTMk.xlsx") 
```

```python
tinsk = pd.read_excel("TINSk.xlsx")
```

```python
incok = pd.read_excel("INCOk.xlsx")
```

```python
print(antmk)
print(tinsk)
print(incok)
```

```python
cor1 = antmk.corr(method='pearson')
cor2 = tinsk.corr(method='pearson')
cor3 = incok.corr(method='pearson')

print(cor1)
print(cor2)
print(cor3)
```

```python
e1 = antmk.filter(['epoch'])
r1 = antmk.filter(['rmse'])

e2 = tinsk.filter(['epoch'])
r2 = tinsk.filter(['rmse'])

e3 = incok.filter(['epoch'])
r3 = incok.filter(['rmse'])
```

```python
plt.title("Korelasi Epoch Terhadap RMSE ANTM")
plt.scatter(e1,r1, color = 'green')
plt.plot(e1,r1,color='red')
plt.ylabel("RMSE")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.title("Korelasi Epoch Terhadap RMSE TINS")
plt.scatter(e2,r2, color = 'green')
plt.plot(e2,r2,color='red')
plt.ylabel("RMSE")
plt.xlabel("Epoch")
plt.show()
```

```python
plt.title("Korelasi Epoch Terhadap RMSE INCO")
plt.scatter(e3,r3, color = 'green')
plt.plot(e3,r3,color='red')
plt.ylabel("RMSE")
plt.xlabel("Epoch")
plt.show()
```
